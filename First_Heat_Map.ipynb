```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Distance_In_Miles_Clean.csv')
```

```python
df
```

```python
# Identifying string values, with all other values being floats.

def convert_to_string(value):
    if isinstance(value, (int, float)):
        return value
    else:
        return str(value)

# Run it through the DataFrame
df_with_strings = df.map(convert_to_string)

# Calculate costs by multiplying numeric values by the Reefer Rate (fridge freight cost in README)
Reefer_Rate = 2.05

# Multiply the numerical values, skipping strings
df_with_cost = df_with_strings.map(lambda x: x if isinstance(x, str) else x * Reefer_Rate)

# Round it only 2 decimal places because its dollar amounts.
df_with_cost_rounded = df_with_cost.round(2)

# Display the DataFrame with updated cost values
print(df_with_cost_rounded)
```

```python
df_with_cost_rounded 
```

```python
# Set Location column as the index
# df_with_cost_rounded.set_index('Location', inplace=True)

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_with_cost_rounded, annot=True, cmap='viridis', fmt=".2f")
plt.title('Distance Between NJ Counties (in miles)')
plt.xlabel('Counties')
plt.ylabel('Counties')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# plt.savefig('Distance_With_Cost.jpg', format='jpg', dpi=300)
```
